```python
# coding:utf-8
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://news.cqjtu.edu.cn/rss.htm"
re = requests.get(url, 'lxml')
re.encoding = "utf-8"
soup = BeautifulSoup(re.text)
news_sorts = {"ywDiv_0":"时政要闻", "JxDiv_0":"教学科研", "JxDiv_1":"招生就业", "JDiv_0":"人文交大", "JDiv_1":"合作交流",
              "JKDiv_0":"信息通知", "JKDiv_1":"学术信息", "JKDiv_2":"招生信息"}
keys = news_sorts.keys()
for key in keys:
    news_titles = soup.select("#" + key + " a")
    print(news_sorts[key] + '\n')
    # news_titles
    for n in news_titles:
        title = n.get("title")
        href = n.get("href")
        data = {
            '标题':title,
            '链接':"http://news.cqjtu.edu.cn/" + href
        }
        print(data)
    print('\n\n')
```
