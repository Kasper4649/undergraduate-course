## 下面批量处理每一年的流感数据

```python
file_name = os.listdir('by_year')                             #获得 by_year 文件夹内的文件名[list 形式]
```

## 将上述方法整合封装为函数

```python
def Preprogress(df, year):
    data_columns = df.iloc[1]
    data_columns[0] = " "
    df.columns = list(data_columns)                            #更改列名
    df.drop(df.columns[0], axis=1, inplace=True)               # 删除第 1 列
    df.drop([0,1,len(df)-1], axis=0, inplace=True)             # 删除头两行和后一行
    df["年份"] = year 
    return(df)                                       
```

```python
def ReadYear(file_name, the_path):
    list = []                                                        # 建立空列表用于存放数据
    for i in range(len(file_name)):                                  # 通过循环遍历读取文件                 
        df = pd.read_csv(the_path + file_name[i], encoding="gbk")    # 读取数据
        year = 2004 + i                                              # 依次累加年份
        Preprogress(df, year)                                        # 进行预处理及其重塑变量
        list.append(df)                                              # 将处理过后的数据框添加到 list 中
        other_data = pd.concat(list)                                 # 使用 concat 合并 list 内的数据
    return(other_data)                                            

total_data = ReadYear(file_name, "by_year/")                         
total_data.tail() 
```

```python
total_data.reset_index(inplace=True, drop=True)             #重新设置索引
total_data.fillna(0, inplace=True)                          #将缺省值设为 0
```

```python
total_data.head()
```

```python
total_data["地区"] = total_data["地区"].apply(lambda x: x.replace(" ", ""))    # 替换文字中的空格
total_data = total_data.loc[total_data["地区"] != "建设兵团"]                  # 删除地区为建设
total_data.loc[total_data['地区'] == '黑龙江', '地区'] = '黑龙江省'             # 将黑龙江替换为黑龙江省
```

```python
people = pd.read_csv("people2.csv" , encoding="gbk") 
people.head()
```

```python
people_columns = list(people.iloc[2])                                       # 用第 2 行作为列名
people.columns.insert(0, " ")
people.columns = people_columns
people.drop([0,1,2,len(people)-1,len(people)-2], axis=0, inplace=True)     # 删除多余的行
people.reset_index(inplace=True, drop=True)                                # 重建索引
people.head()
```

```python
## 统一地区名的格式
people.loc[people['地区'] == '内蒙古自治区', '地区'] = '内蒙古'
people.loc[people['地区'] == '广西壮族自治区', '地区'] = '广西'
people.loc[people['地区'] == '西藏自治区', '地区'] = '西藏'
people.loc[people['地区'] == '宁夏回族自治区', '地区'] = '宁夏'
people.loc[people['地区'] == '新疆维吾尔自治区', '地区'] = '新疆'
```

```python
people.head()
```

```python
peo_name = list(people.columns)                                                        # 获取 people 的变量名
peo_name.remove("地区")                                                                # 去除地区变量，得到年份数据
change_people = pd.melt(people, id_vars=["地区"], value_vars=peo_name, \
                        var_name="年份", value_name="总人口数")                        # 通过 melt 重塑数据
change_people["年份"] = change_people["年份"].apply(lambda x: re.findall("\d+", x)[0]) # 去除年份的“年”字
change_people["年份"] = change_people["年份"].astype(np.int)                           # 将年份转换为数值形式
change_people.head()
```

```python
result = pd.merge(total_data, change_people, on=['年份', '地区'])
```

```python
result.head()
```

```python
change_list = ['发病率','死亡率','总人口数','发病数','死亡数']
result[change_list] = result[change_list].apply(pd.to_numeric)
result.head()
```
